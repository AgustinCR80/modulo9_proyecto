### Modificar los valores de la columna 'Tipo Carga':

```python
# Clasificar los valores de potencia en tipos de velocidad de carga

def tipo_velocidad_carga (potencia):
    """
    Función para clasificar potencias.

    Parámetros:
    - numero (int): Número que se va a categorizar.

    Retorna:
    - str: Categoría a la que pertenece la potencia.
    """
    if potencia == 0:
        return 'ND 0kW'
    elif 1 <= potencia <= 4:
        return 'Carga Lenta (AC)'
    elif 5 <= potencia <= 21:
        return 'Carga Semirrápida (AC)'
    elif 22 <= potencia <= 99:
        return 'Carga Rápida (DC)'
    elif potencia >= 100:
        return 'Carga Ultrarrápida (DC)'
    else:
        return 'No clasificado'
```

```python
# Aplicar la función a la columna
cargadores['Tipo Carga'] = cargadores['Tipo Carga'].apply(tipo_velocidad_carga)
```

# Visualizar estadísticas

## Matriculaciones

### Matriculaciones por provincia en cada año

```python
# Agrupar por Provincia y Año, sumando las Unidades
provincia_año = matriculaciones.groupby(["Provincia", "Año"])["Unidades"].sum().reset_index().T
provincia_año
```

```python
# Inicializar una figura
plt.figure(figsize=(10, 6))

# Customizar gráfico
sns.barplot(data=provincia_año.T, x='Año', y='Unidades', hue='Provincia')
plt.title('Provincia-Año')
plt.xlabel('Año')
plt.ylabel('Unidades')
plt.tight_layout()
plt.grid(alpha= 0.4)  # Mostar una cuadrícula e intensidad de las líneas
plt.xticks(rotation=45) # Rotación etiquetas

# Mostrar la figura
plt.show()

display(provincia_año)
```

##### Hay que tener en cuenta que 2025 no está completo, pero en el resto de los años se aprecia una tendencia ascentente en todas las provincias.

### Cantidad de vehículos matriculados por tecnología

```python
# Agrupar por Tecnología y sumar las unidades
categorias_tecnologia = matriculaciones.groupby('Tecnologia', as_index=False)['Unidades'].sum().sort_values(by='Unidades', ascending=False)
categorias_tecnologia
```

```python
# Inicializar una figura
plt.figure(figsize=(10, 6))

# Customizar gráfico
sns.barplot(x='Tecnologia', y='Unidades', data=categorias_tecnologia, hue='Tecnologia')
plt.title(f'{'Tecnologia'}-{'Unidades'}')
plt.xlabel(f'{'Tecnologia'}')
plt.ylabel('Unidades')
plt.tight_layout()
plt.grid(alpha= 0.4)  # Mostar una cuadrícula e intensidad de las líneas
plt.xticks(rotation=45) # Rotación etiquetas

# Mostrar la figura
plt.show()

display(categorias_tecnologia.T)
```
